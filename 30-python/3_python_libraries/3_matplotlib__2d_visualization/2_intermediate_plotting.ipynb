## Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 1, 10)
y = x ** 2
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 3))
plt.tight_layout()
# for ax in axes:
#     ax.plot(x, y)
axes[0].plot(x, y)
axes[1].plot(y, x)
```

### Figure Size and DPI

```python
fig = plt.figure(figsize=(8, 2))
axis = fig.add_axes([0,0,1,1,])
axis.plot(x, x ** 3, label='X Cute')
axis.plot(y, y ** 2, label='Y Square')

axis.legend(loc=(0.05,0.7)) # loc=0 | loc='right'
```

### Save Figure

```python
fig.savefig("figure_example.jpeg", dpi=200)
```
