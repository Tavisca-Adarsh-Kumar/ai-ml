## Matplotlib - Advanced Plotting

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 10)
y = x ** 2
```

```python

fig = plt.figure()

ax = fig.add_axes([0.1,0.1,0.2,0.2])
ax.plot(x, y, color='red', lw=0.5, ls='-.', marker='o', ms=5, mfc='green', mew=2, alpha=0.5) # lw: linewidth | ls: linestyle='steps'
ax.set_xlim(0, 2)
ax.set_ylim(0, 2)
```

### Special Plot Types

```python
plt.scatter(x, y)
```
