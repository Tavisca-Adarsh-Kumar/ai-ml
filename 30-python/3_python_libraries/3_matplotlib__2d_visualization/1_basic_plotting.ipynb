## Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
x
```

```python
plt.plot(x, y)
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.title('Two Dimensional Plot')
plt.show()
```

### Multi Plot

```python
plt.subplot(1, 2, 1) # row, col, plot_number
plt.plot(x, y, 'r')  # r: red color

plt.subplot(1, 2, 2) # row, col, plot_number
plt.plot(y, x, 'b') # b: blue color
```

### Object Oriented

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.7,0.7])
axes2 = fig.add_axes([0.2,0.4,0.4,0.4]) # (x, y, width, height)

axes1.plot(x, y)
axes1.set_title('Parent Title')
axes2.plot(y, x)
```
