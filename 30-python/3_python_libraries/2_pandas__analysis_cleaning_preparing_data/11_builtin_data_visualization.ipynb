## Pandas - Built-in Visualization

```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df1 = pd.read_csv("../../50_bootcamp_examples/07_pandas__built_in_data_viz/df1")
```

```python
df1.head(4)
```

```python
df1['A'].hist(bins=40)
```

```python
df1['A'].plot(kind='hist', bins=40)
```

```python
df1['A'].plot.hist()
```

```python
df2 = pd.read_csv("../../50_bootcamp_examples/07_pandas__built_in_data_viz/df2")
df2.head()
```

```python
df2.plot.area(alpha=0.4)
```

```python
df2.plot.bar(stacked=True)
```

```python
df1['A'].plot.line(x=df1.index, y='B', figsize=(10,4), lw=1)
```

```python
df1.plot.scatter(x='A', y='B', c='C', cmap='coolwarm', s=df1['C']*100)
```

```python
df2.plot.box()
```

```python
df2.plot.hexbin(x='a', y='b', gridsize=25)
```

```python
df2.plot.density()
```
