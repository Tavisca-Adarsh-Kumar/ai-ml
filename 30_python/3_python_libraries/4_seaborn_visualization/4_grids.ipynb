## Seaborn - Grids

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
iris['species'].unique()
```

```python
pg = sns.PairGrid(iris)
pg.map_diag(sns.displot)
pg.map_upper(plt.scatter)
pg.map_lower(sns.kdeplot)
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
pg = sns.FacetGrid(data=tips, col='time', row='smoker')
pg.map(sns.displot, 'total_bill', 'tip')
```
